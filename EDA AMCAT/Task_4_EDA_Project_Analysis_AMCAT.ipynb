```python
cat_col.drop("DOL",axis=1,inplace=True)
```

```python
def cat_column(cat_col):
    for col_name in cat_col:
        print(col_name)
        print(cat_col[col_name].agg(["unique","nunique","count"]))
        print(f"         value counts:\n{cat_col[col_name].value_counts()}")
        print("\n")
```

```python
cat_column(cat_col)
```

```python
num_col.drop("ID",axis =1,inplace = True)
```

```python
def num_column(num_col):
    print("               Non Visualization                     ")
    for col_name in num_col:
        print(col_name)
        print(num_col[col_name].agg(["max","min","mean","median","std","skew","kurt"]))
        print("\n")
```

```python
num_column(num_col)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def box_plot(num_col):
    for col_name in num_col:
        plt.title(f"Box Plot for {col_name}")
        plt.boxplot(num_col[col_name])
        plt.xlabel(col_name)
        plt.xticks(rotation = 0)
        plt.show()
```

```python
box_plot(num_col)
```
